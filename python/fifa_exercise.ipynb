```python
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('WorldCupMatches.csv')
df['Total Goals'] = df['Home Team Goals'] + df['Away Team Goals']
```

```python
sns.set_style('whitegrid')
sns.set_context('poster', font_scale = 0.4)
fig, ax = plt.subplots(figsize = (12, 7))
ax = sns.barplot(data = df, x ='Year', y ='Total Goals')
ax.set_title('Average Number of Goals Scored in Matches by Year')
plt.show()
```

```python
df_goals = pd.read_csv('goals.csv')
```

```python
sns.set_context('notebook', font_scale=1.25)
fig2, ax2 = plt.subplots(figsize=(12, 7))
ax2 = sns.boxplot(data = df_goals, x = 'year', y = 'goals', palette = 'Spectral')
ax2.set_title('Average Range of Goals Scored in Matches by Year')
plt.show()
```
