### Update 'damages' list entries:

```python
updated_damages = []
def remove_suffix_and_multiply(damage_list):
    for cost in damage_list:
        if cost == 'Damages not recorded':
            updated_damages.append(cost)
        elif cost[-1] == 'M':
            updated_damages.append(float(cost.strip("M")) * 1000000)
        elif cost[-1] == 'B':
            updated_damages.append(float(cost.strip("B")) * 1000000000)
    return updated_damages
remove_suffix_and_multiply(damages) 
```

### Creating a hurricane dictionary:

```python
def create_hurricane_dictionary(names, months, years, max_sustained_winds, areas_affected, updated_damages, deaths):
    hurricane_dictionary = {}
    num_hurricanes = len(names)
    for index in range(num_hurricanes):
        hurricane_dictionary[names[index]] = {"Name": names[index], \
                                              "Month": months[index], \
                                              "Year": years[index], \
                                              "Max Sustained Wind": max_sustained_winds[index], \
                                              "Areas Affected": areas_affected[index], \
                                              "Damage": updated_damages[index], \
                                              "Deaths": deaths[index]}
    return hurricane_dictionary
create_hurricane_dictionary(names, months, years, max_sustained_winds, areas_affected, updated_damages, deaths)
```

### Alter dictionary to make 'Year' the key:

```python
def year_cane_dictionary(dictionary):
    hurricane_year = []
    for hurricane in dictionary:
        current_year = dictionary[hurricane]['Year']
        current_cane = dictionary[hurricane]
        year_hurricane_dictionary[current_year] = current_cane
    return year_hurricane_dictionary
year_cane_dictionary(hurricane_dictionary)
```

### Create a dictionary of areas hit the number of times hit:

```python
# write your count affected areas function here:
def affected(dictionary):
    areas_affected_list = []
    for hurricane in dictionary:
        areas_affected_cane = hurricane_dictionary[hurricane]['Areas Affected']
        for place in areas_affected_cane:
            areas_affected_list.append(place)

    unique_area_list = []
    for place in areas_affected_list:
        if place not in unique_area_list:
            unique_area_list.append(place)
    
    unique_area_count_list = []    
    for place in unique_area_list:
        count = areas_affected_list.count(place)
        unique_area_count_list.append(count)
    
    location_count_dictionary = {location: count for location, count in zip(unique_area_list, unique_area_count_list)}
    return location_count_dictionary
affected_count_dictionary = affected(hurricane_dictionary)
print(affected_count_dictionary)
```

### Find which area was hit the most and how many times:

```python
def most_affected(dictionary):
    top_hit_area = ""
    top_hit_area_count = 0
    for location in dictionary:
        if dictionary[location] > top_hit_area_count:
            top_hit_area = location
            top_hit_area_count = dictionary[location]
    return print("{} was affected the most being hit {} time(s)!"\
                 .format(top_hit_area, top_hit_area_count))
most_affected(location_count_dictionary)
```

### Find which hurricane was the deadliest and number of associated deaths:

```python
def most_deaths(dictionary):
    highest_death_count = 0
    highest_death_hurricane = ""
    for hurricane in dictionary: 
        if dictionary[hurricane]['Deaths'] > highest_death_count:
            highest_death_count = dictionary[hurricane]['Deaths']
            highest_death_hurricane = hurricane
    return print("{} was the deadliest hurricane with {} deaths!"\
                 .format(highest_death_hurricane, highest_death_count))
most_deaths(hurricane_dictionary)
```

### Create a dictionary of hurricanes categorized by death rate:

```python
def mortality(dictionary):
    mortality_scale = {0: 0, 1: 100, 2: 500, 3: 1000, 4: 10000}
    hurricanes_by_mortality = {0: [], 1: [], 2: [], 3: [], 4: [], 5: []}
    for hurricane in dictionary:
        if dictionary[hurricane]['Deaths'] > mortality_scale[4]:
            hurricanes_by_mortality[5].append(hurricane)
        elif dictionary[hurricane]['Deaths'] > mortality_scale[3]:
            hurricanes_by_mortality[4].append(hurricane)
        elif dictionary[hurricane]['Deaths'] > mortality_scale[2]:
            hurricanes_by_mortality[3].append(hurricane)
        elif dictionary[hurricane]['Deaths'] > mortality_scale[1]:
            hurricanes_by_mortality[2].append(hurricane)
        elif dictionary[hurricane]['Deaths'] > mortality_scale[0]:
            hurricanes_by_mortality[1].append(hurricane)
        else: 
            hurricanes_by_mortality[0].append(hurricane)
    return(hurricanes_by_mortality)
mortality(hurricane_dictionary)
```

### Find which hurricane caused the most damage and the associated costs:

```python
# write your greatest damage function here:
def greatest_damage(dictionary):
    greatest_damage_cost = 0
    greatest_damage_hurricane = ''
    for hurricane in dictionary:
        try:
            if dictionary[hurricane]['Damage'] > greatest_damage_cost:
                greatest_damage_cost = dictionary[hurricane]['Damage']
                greatest_damage_hurricane = hurricane
        except TypeError:
            continue
    return print("{} caused the most damage at a cost of ${}!".format(greatest_damage_hurricane, greatest_damage_cost))

greatest_damage(hurricane_dictionary)
```

```python
# write your catgeorize by damage function here:
def damage(dictionary):
    damage_scale = {0: 0, 1: 100000000, 2: 1000000000, 3: 10000000000, 4: 50000000000}
    hurricanes_by_damage = {0: [], 1: [], 2: [], 3: [], 4: [], 5: [], "Unknown": []}
    for hurricane in dictionary:
        try:
            if dictionary[hurricane]['Damage'] > damage_scale[4]:
                hurricanes_by_damage[5].append(hurricane)
            elif dictionary[hurricane]['Damage'] > damage_scale[3]:
                hurricanes_by_damage[4].append(hurricane)
            elif dictionary[hurricane]['Damage'] > damage_scale[2]:
                hurricanes_by_damage[3].append(hurricane)
            elif dictionary[hurricane]['Damage'] > damage_scale[1]:
                hurricanes_by_damage[2].append(hurricane)
            elif dictionary[hurricane]['Damage'] > damage_scale[0]:
                hurricanes_by_damage[1].append(hurricane)
            else:
                hurricanes_by_damage[0].append(hurricane)
        except TypeError:
            hurricanes_by_damage["Unknown"].append(hurricane)
    return hurricanes_by_damage
damage(hurricane_dictionary)



```
